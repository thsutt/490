```python
from matplotlib import pyplot as plt
import numpy as np


arr_labels = ["accuracy", "recall", "auc"]

text = [0.58, 0.58, 0.56]

img = [0.56, 0.45, 0.59]

comb = [0.65, 0.7, 0.64]

fig, ax = plt.subplots()

x = np.arange(3)

# Plot the bars
ax.bar(x - 0.2, text, width=0.2, capsize=5, label='Text Only', bottom=0)
ax.bar(x, img, width=0.2, capsize=5, label='Image Only', bottom=0)
ax.bar(x + 0.2, comb, width=0.2,capsize=5, label='Combined', bottom=0)



# Set the x-axis ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(arr_labels)

# Set labels and title
# ax.set_xlabel('Stats')
ax.set_ylabel('Value')
ax.set_title('Learning Metrics for All Three Classifiers')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
```
