```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from collections import Counter
```

```python
data = pd.read_csv('mdb.tsv', sep=' ')
```

```python
B = data[(data["severity"] == 'B')]
M = data[(data["severity"] == 'M')]
```

```python
fig, ax1 = plt.subplots(1, 1)
fig.suptitle('Class of Benign and Malignant Growths')



category_counts = Counter(B["class"])
category_counts_M = Counter(M["class"])
categories = list(category_counts.keys())
B_counts = list(category_counts.values())
M_counts = list(category_counts_M.values())

x = np.arange(len(categories))
bar_width = 0.35

ax1.bar(x - bar_width/2, B_counts, bar_width, label='Benign', color='blue')
ax1.bar(x + bar_width/2, M_counts, bar_width, label='Malignant', color='red')

ax1.set_xlabel('Class')
ax1.set_ylabel('Count')
ax1.set_xticks(x)
ax1.set_xticklabels(categories)
ax1.legend()
```

```python
fig, ax1 = plt.subplots(1, 1)
plt.subplots_adjust(left=-0.1)
fig.suptitle("Radius of Benign and Malignant Growths")

ax1.set_xlabel('Radius')
ax1.set_ylabel('Count')
ax1.hist([B["radius"], M["radius"]], color=["blue", "red"], label=["Benign", "malignant"])
ax1.legend()
```

```python
norm_count = len(data.loc[data["class"] == "NORM"])
benign_count = len(data.loc[data["severity"] == "B"])
malignant_count = len(data.loc[data["severity"] == "M"])
```

```python
labels = ['Normal', 'Benign', 'Malignant']
sizes = [norm_count, benign_count, malignant_count]  # Percentages

# Create a pie chart
plt.pie(sizes, labels=labels, colors=["lightgray", "blue", "red"], autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('')

plt.show()
```
