```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize  # Make sure to install nltk and download punkt tokenizer data
import nltk
# nltk.download('punkt')
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('mdb.tsv', sep=' ', dtype='str')
```

```python
circ_data = data[data.loc[:, 'class'] == 'CIRC']
```

```python
B = data[(data["severity"] == 'B')]
M = data[(data["severity"] == 'M')]
NORM = data[(data["class"] == 'NORM')]
```

```python
B_circ = circ_data[(circ_data["severity"] == 'B')]
M_circ = circ_data[(circ_data["severity"] == 'M')]
```

```python
# B_Sentences = [' '.join(map(str, row)) for row in B[['tissue', 'class', 'x', 'y', 'radius']].values.tolist()]
# M_Sentences = [' '.join(map(str, row)) for row in M[['tissue', 'class','x', 'y', 'radius']].values.tolist()]
```

```python
def convert_item(val):
    conversion_dict = {
        'F': 'Fatty',
        'G': 'Fatty-glandular',
        'D': 'Dense-glandular',
        'CALC': 'Calcification',
        'CIRC': 'Well-defined/circumscribed masses',
        'SPIC': 'Spiculated masses',
        'MISC': 'Other, ill-defined masses',
        'ARCH': 'Architectural distortion',
        'ASYM': 'Asymmetry',
        'NORM': 'Normal',
    }

    return conversion_dict.get(val, str(val))

#Make norm sentences extra long so 
def convert_item_norm(val):
    conversion_dict = {
            'F': 'Fatty Fatty',
            'G': 'Fatty-glandular Fatty-glandular',
            'D': 'Dense-glandular Dense-glandular',
        }
   
    return conversion_dict.get(val, str(val))
```

```python
B_sentences = [' '.join(map(convert_item, row)) for row in B[['tissue', 'class']].values.tolist()]
M_sentences = [' '.join(map(convert_item, row)) for row in M[['tissue', 'class']].values.tolist()]
NORM_sentences = [' '.join(map(convert_item_norm, row)) for row in NORM[['tissue']].values.tolist()]
NORM_sentences
```

```python
B_c_sentences = [' '.join(map(convert_item, row)) for row in B_circ[['tissue']].values.tolist()]
M_c_sentences = [' '.join(map(convert_item, row)) for row in M_circ[['tissue']].values.tolist()]
```

```python
MB_sentences = B_sentences + M_sentences
MB_c_sentences = B_c_sentences + M_c_sentences
All_sentences = MB_sentences + NORM_sentences
```

```python
# Function to get the vector for a description
def get_description_vector_avg(description, model):
    words = word_tokenize(description.lower())
    
    # Filter out words that are not in the vocabulary
    words_in_vocab = [word for word in words if word in model.wv.key_to_index]
    if words_in_vocab:
        # Calculate the average vector
        return sum(model.wv[word] for word in words_in_vocab) / len(words_in_vocab)
    else:
        # If no words are in the vocabulary, return a vector of zeros
        return [0] * model.vector_size
```

```python
# Function to get the vector for a description
def get_description_vector_sum(description, model):
    words = word_tokenize(description.lower())
    
    # Filter out words that are not in the vocabulary
    words_in_vocab = [word for word in words if word in model.wv.key_to_index]
    if words_in_vocab:
        # Calculate the sum vector
        return sum(model.wv[word] for word in words_in_vocab)
    else:
        # If no words are in the vocabulary, return a vector of zeros
        return [0] * model.vector_size
```

```python
sentences = All_sentences
get_vec = get_description_vector_avg

# Tokenize the text into words
tokenized_text = [word_tokenize(desc.lower()) for desc in sentences]  # Convert to lowercase for consistency

# Define and train Word2Vec model
embedding_dim = 100  # You can adjust the dimensionality of the embeddings
model = Word2Vec(sentences= tokenized_text, vector_size=embedding_dim, window=5, min_count=1, workers=4)

# # Save the model (optional)
# model.save("word2vec_model.bin")


```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
def compress_count(data):
    #data is an array of pairs (stored as arrays)
    compressed = {}
    for element in data:
        key = tuple(element)
        if key not in compressed:
            compressed[key] = 1
        else:
            compressed[key] += 1
    return compressed
```

```python
def get_compressed_arrays(data):
    compressed = compress_count(data)
    z = list(compressed.values())
    pairs = [np.array(t) for t in compressed.keys()]
    x, y = [list(t) for t in zip(*pairs)]
    return(x,y,z)
```

```python
desc_vectors1 = [get_description_vector_sum(desc, model) for desc in B_sentences] 
data1_label = "Normal"

desc_vectors2 = [get_description_vector_sum(desc, model) for desc in M_sentences]
data2_label = "Abnormal"
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score

aggregate_data = desc_vectors1 + desc_vectors2
data_labels = [0] * len(desc_vectors1) + [1] * len(desc_vectors2)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(aggregate_data, data_labels, test_size=0.2, random_state=42)

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)
# report = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy}")
print(f"Recall: {recall}")
print(f"AUC: {auc}")
```
